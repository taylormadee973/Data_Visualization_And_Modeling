# Lab Instructions

You are part of a data team for a financial institution that uses an AI model to determine who will and will not be granted a loan.  Recently, there have been complaints from customers that the model unfairly discriminates based on the applicant's age and sex. A court has ordered that the financial institution must provide information about how the model works.  

Data from recent loan applications is given in `loan_data.csv`.  The data dictionary is below:

| Feature              | Description | Data Type |
| :---------------- | :------: | ----: |
|person_age|Age of the person|Float
|person_gender|Gender of the person|Categorical
|person_education|Highest education level|Categorical
|person_income|Annual income|Float
|person_emp_exp|Years of employment experience|Integer
|person_home_ownership|Home ownership status (e.g., rent, own, mortgage)|Categorical
|loan_amnt|Loan amount requested|Float
|loan_intent|Purpose of the loan|Categorical
|loan_int_rate|Loan interest rate|Float
|loan_percent_income|Loan amount as a percentage of annual income|Float
|cb_person_cred_hist_length|Length of credit history in years|Float
|credit_score|Credit score of the person|Integer
|previous_loan_defaults_on_file|Indicator of previous loan defaults|Categorical
|loan_status (target variable)|Loan approval status: 1 = approved; 0 = rejected|Integer

Visualize the relationship between age, sex and 5 other features with `loan_status`.  Which features seem to drive the model's decision to grant a loan?  Which seem unrelated?  Should the financial institution be concerned about potential legal trouble?  Make sure to justify your answers with specific references to your visualizations.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("loan_data.csv")

df.head()
```

```python
plt.figure(figsize=(15,10))

# Age vs Loan Status
plt.subplot(3,3,1)
sns.histplot(x='person_age', data=df, hue='loan_status', multiple='stack', bins=20, palette='coolwarm')
plt.title("Age vs Loan Approval")
plt.legend(title='Status', labels=['Rejected','Approved'])

# Gender vs Loan Status
plt.subplot(3,3,2)
sns.countplot(x='person_gender', hue='loan_status', data=df, palette='coolwarm')
plt.title("Gender vs Loan Approval")
plt.legend(title='Status', labels=['Rejected','Approved'])

# Education vs Loan Status
plt.subplot(3,3,3)
sns.countplot(x='person_education', hue='loan_status', data=df, palette='coolwarm')
plt.title("Education vs Loan Approval")
plt.xticks(rotation=45)
plt.legend(title='Status', labels=['Rejected','Approved'])

# Income vs Loan Status
plt.subplot(3,3,4)
sns.boxplot(x='loan_status', y='person_income', data=df, palette='coolwarm')
plt.title("Income vs Loan Approval")
plt.xlabel("Loan Status (0=Rejected, 1=Approved)")

# Employment Experience vs Loan Status
plt.subplot(3,3,5)
sns.boxplot(x='loan_status', y='person_emp_exp', data=df, palette='coolwarm')
plt.title("Employment Experience vs Loan Approval")
plt.xlabel("Loan Status (0=Rejected, 1=Approved)")

# Credit Score vs Loan Status
plt.subplot(3,3,6)
sns.boxplot(x='loan_status', y='credit_score', data=df, palette='coolwarm')
plt.title("Credit Score vs Loan Approval")
plt.xlabel("Loan Status (0=Rejected, 1=Approved)")

# Loan Intent vs Loan Status
plt.subplot(3,3,7)
sns.countplot(x='loan_intent', hue='loan_status', data=df, palette='coolwarm')
plt.title("Loan Intent vs Loan Approval")
plt.xticks(rotation=45)
plt.legend(title='Status', labels=['Rejected','Approved'])

# Home Ownership vs Loan Status
plt.subplot(3,3,8)
sns.countplot(x='person_home_ownership', hue='loan_status', data=df, palette='coolwarm')
plt.title("Home Ownership vs Loan Approval")
plt.legend(title='Status', labels=['Rejected','Approved'])

plt.tight_layout()
plt.show()
```

# Key Takeaways
- Younger applicants were rejected more often than middle-aged ones.  
- Gender differences show unequal approval rates.  
- Higher income and credit scores increased approval odds.  
- More employment experience helped slightly.  
- Education and home ownership also influenced approvals.  
- Loan intent (purpose) seemed less related to approval.


# Conclusions on Fairness and Ethics

Most related to approval:
- Credit score, income, and employment experience.

Least related:
- Loan intent and education.

Possible bias:
- Age and gender appear to influence loan decisions.

Ethical concern:
- If the model uses age or gender, it may violate fair-lending laws.
- The institution should retrain the model using only financial variables.
