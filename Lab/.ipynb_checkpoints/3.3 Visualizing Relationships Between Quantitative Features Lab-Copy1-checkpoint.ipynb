# Lab Instructions

You are working as an analyst for a consulting agency that has been hired to study and improve productivity at a mid-sized company.  Morale is very low and, after speaking with employees, several common complaints kept coming up:

* New employees complained that more tenured employees were lazy and unproductive.
* More tenured employees complained that the newer employees didn't focus and were therefore unproductive.
* Managers were worried that remote workers might be less productive.
* In-person employees found it impossible to focus and be productive when their desk was in the open plan area.
* Many people said they had so much administrative work to do they could only be truly productive if they worked overtime.
* Everyone complained that there were too many meetings.

The consulting agency carefully collected data on a number of factors including number of years at the company, focus (higher is better), productivity (higher is better), department, meetings per week, and desk location can be found in Employee_productivity.csv.

1. Determine which feature - years at the company, hours worked, or focus - has the strongest linear relationship with productivity.  Show a visualization and calculate R-squared for all three relationships to justify your answer.

2. Build on that linear regression model to determine what other featuers also impact productivity. Using both visualzations and R-squared values, build a model that can predicts employee productivity at this company. Evaluate each feature in the dataset.

3. Write a short (a few sentences) summary of how the different factors in the data relate to employee productivity.  What recommendations would you make so that employees can be as productive as possible?



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv("Employee_productivity.csv")

df.head()
print(df.info())
print(df.describe())
```

```python
features = ['YearsAtCompany', 'HoursWorked', 'FocusScore']
target = 'ProductivityScore'
r2_scores = {}

for feature in features:
    plt.figure(figsize=(10,6))
    sns.regplot(x=feature, y=target, data=df, line_kws={'color':'red'})
    plt.title(f'{feature} vs ProductivityScore')
    plt.xlabel(feature)
    plt.ylabel('ProductivityScore')
    plt.grid(True)
    plt.show()

    X = df[[feature]]
    y = df[target]
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    r2_scores[feature] = r2

r2_scores
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)

X = df_encoded.drop(columns=['ProductivityScore'])
y = df_encoded['ProductivityScore']

model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
r2_total = r2_score(y, y_pred)
print("Overall Model R:", round(r2_total, 2))

coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
}).sort_values(by='Coefficient', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='Coefficient', y='Feature', data=coefficients)
plt.title('Feature Importance for Predicting Productivity')
plt.grid(True)
plt.show()

coefficients.head(10)
```

```python
import numpy as np

# rebuild log columns safely
df['log_FocusScore'] = np.log(df['FocusScore'] + 1)
df['log_Productivity'] = np.log(df['ProductivityScore'] + 1)

# replace infinite or NaN values caused by log
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(subset=['log_FocusScore', 'log_Productivity'], inplace=True)

# visualize log-log relationship
plt.figure(figsize=(10,6))
sns.regplot(x='log_FocusScore', y='log_Productivity', data=df, line_kws={'color':'red'})
plt.title('Log(FocusScore) vs Log(ProductivityScore)')
plt.xlabel('Log(FocusScore)')
plt.ylabel('Log(ProductivityScore)')
plt.grid(True)
plt.show()

# compute R for cleaned log-log data
X = df[['log_FocusScore']]
y = df['log_Productivity']

model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
print("R (Log-Log Relationship):", round(r2_score(y, y_pred), 2))
```

## Summary and Recommendations

**Findings:**  
After analyzing the data, the results show that **FocusScore** has the strongest connection to productivity (**R  0.68**). Employees who can stay focused tend to be much more productive overall.  
**HoursWorked** showed little to no correlation with productivity, indicating that longer hours dont necessarily lead to better results.  
**YearsAtCompany** had the weakest relationship, suggesting that experience alone doesnt guarantee higher productivity.  
When all variables were considered together (**overall R  0.69**), the model revealed that focus, meeting frequency, and workspace setup had the biggest impact on productivity.

**Insights:**  
Employees who work remotely or have private offices generally performed better than those in open workspaces.  
Too many meetings or excessive administrative tasks reduced focus and productivity, aligning with employee feedback about distractions and workload.

**Recommendations:**  
1. Reduce unnecessary meetings to allow more uninterrupted focus time.  
2. Offer remote or quieter workspace options for employees who need fewer distractions.  
3. Simplify or automate repetitive administrative tasks.  
4. Encourage time management and focus-building strategies to sustain productivity.

Overall, employees are most productive when they can focus deeply in an environment that minimizes interruptions.
