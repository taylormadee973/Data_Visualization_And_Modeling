## Dataset Description
This lab uses the **Historical Hourly Weather Data (2012 ‚Äì 2017)** dataset from Kaggle.  
It includes hourly weather observations for multiple global cities such as temperature, humidity, and wind speed.

**Features**
- `datetime` ‚Äì date and time of observation  
- `city` ‚Äì city name  
- `temperature` (¬∞F) ‚Äì hourly average temperature converted from Kelvin  
- `humidity` (%) ‚Äì hourly average humidity  

This dataset allows analysis of **temporal trends** (how weather changes over time) and **spatial variation** (differences between cities).


```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("temperature.csv")

# Convert from wide to long format
df_long = df.melt(id_vars='datetime', var_name='city', value_name='temperature')

# Convert Kelvin to Fahrenheit
df_long['temperature_F'] = (df_long['temperature'] - 273.15) * 9/5 + 32
df_long['datetime'] = pd.to_datetime(df_long['datetime'])
df_long.dropna(inplace=True)

cities = ['Los Angeles', 'New York', 'Chicago', 'Miami', 'Seattle']
df_long = df_long[df_long['city'].isin(cities)]

df_long = df_long.groupby(['city', pd.Grouper(key='datetime', freq='ME')])['temperature_F'].mean().reset_index()


print(df_long['city'].unique())
df_long.head()
```

```python
fig = px.line(
    df_long,
    x='datetime',
    y='temperature_F',
    color='city',
    title='üå°Ô∏è Monthly Average Temperature Trends (2012‚Äì2017)',
    labels={'datetime': 'Date', 'temperature_F': 'Temperature (¬∞F)'}
)
fig.update_layout(height=500)
fig.show()
```

### Temperature Trends Analysis
Temperature patterns differ clearly by region.  
Northern cities like **Chicago** and **New York** show cold winters and hot summers,  
while coastal and southern cities like **Los Angeles** and **Miami** remain warmer year-round.  
Seattle maintains cooler, more stable temperatures due to its coastal climate.


```python
df_hum = pd.read_csv("humidity.csv")
df_hum = df_hum.melt(id_vars='datetime', var_name='city', value_name='humidity')
df_hum['datetime'] = pd.to_datetime(df_hum['datetime'])
df_hum = df_hum[df_hum['city'].isin(cities)]

df_hum = df_hum.groupby(['city', pd.Grouper(key='datetime', freq='ME')])['humidity'].mean().reset_index()

# Merge temperature + humidity
merged = pd.merge(df_long, df_hum, on=['datetime', 'city'], how='inner')

fig = px.scatter(
    merged,
    x='temperature_F',
    y='humidity',
    color='city',
    animation_frame=merged['datetime'].dt.strftime('%Y-%m'),
    hover_name='city',
    title='üíß Temperature vs Humidity by U.S. City (2012‚Äì2017)',
    labels={'temperature_F': 'Temperature (¬∞F)', 'humidity': 'Humidity (%)'}
)
fig.update_layout(height=500, xaxis_range=[30, 110], yaxis_range=[0, 100])
fig.show()
```

### Temperature vs Humidity Relationship
This chart shows an inverse relationship between **temperature** and **humidity**.  
As temperatures rise, humidity generally decreases ‚Äî especially in **Los Angeles** and **Chicago**.  
**Miami** stands out for staying humid even at higher temperatures,  
while **Seattle** maintains moderate humidity levels year-round.


```python
coords = {
    'Los Angeles': (34.05, -118.25),
    'New York': (40.71, -74.01),
    'Chicago': (41.88, -87.63),
    'Miami': (25.76, -80.19),
    'Seattle': (47.61, -122.33)
}

df_long['lat'] = df_long['city'].map(lambda c: coords[c][0])
df_long['lon'] = df_long['city'].map(lambda c: coords[c][1])

fig = px.scatter_geo(
    df_long,
    lat='lat',
    lon='lon',
    color='temperature_F',
    hover_name='city',
    animation_frame=df_long['datetime'].dt.strftime('%Y-%m'),
    color_continuous_scale='Turbo',
    title='üó∫Ô∏è Average Monthly Temperatures Across U.S. Cities (2012‚Äì2017)',
    labels={'temperature_F': 'Temperature (¬∞F)'}
)
fig.update_layout(height=500, geo_scope='usa')
fig.show()
```

### Geographic Temperature Distribution
This map highlights how temperature varies across U.S. cities.  
Southern cities like **Miami** and **Los Angeles** remain consistently warm,  
while northern cities such as **Chicago** and **Seattle** cool off significantly in winter.  
Seasonal shifts move gradually from north to south as the months progress.
