```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("spotify-2023.csv", encoding="ISO-8859-1")

df.head()
```

```python
# BPM distribution
sns.histplot(df["bpm"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of BPM")
plt.show()

# Key distribution
sns.countplot(x="key", data=df, order=df["key"].value_counts().index, palette="viridis")
plt.title("Distribution of Key")
plt.show()

# Mode distribution
sns.countplot(x="mode", data=df, palette="Set2")
plt.title("Distribution of Mode (Major/Minor)")
plt.show()

# Danceability, Energy, Speechiness, Acousticness
features = ["danceability_%", "energy_%", "speechiness_%", "acousticness_%"]

for feature in features:
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

Distribution Summary:

Most songs have a BPM between 110–130, showing that mid-tempo songs are the most popular.

The most common keys are A and C#, which are common in pop and dance tracks.

Most songs are in Major mode, but there are some hits in Minor, which gives a more emotional tone.

Danceability usually falls between 60–80%, meaning most popular songs are easy to move to.

Energy typically ranges between 60–80%, showing that hits are lively but not overly intense.

Speechiness tends to be under 10%, which means most songs are sung rather than spoken.

Acousticness is usually under 30%, meaning most hits are produced or electronic rather than purely acoustic.

| Feature          | Ideal Range / Value |
| ---------------- | ------------------- |
| **BPM**          | 120–130             |
| **Key**          | A or C#             |
| **Mode**         | Major               |
| **Danceability** | 70–85%              |
| **Energy**       | 60–80%              |
| **Speechiness**  | < 10%               |
| **Acousticness** | < 30%               |

These values represent the most common features among top songs in the dataset. A song that falls within these ranges would likely have a similar feel and energy as other popular tracks.

```python
filtered_df = df[
    (df["bpm"] >= 120) & (df["bpm"] <= 130) &
    (df["key"].isin(["A", "C#"])) &
    (df["mode"] == "Major") &
    (df["danceability_%"] >= 70) & (df["danceability_%"] <= 85) &
    (df["energy_%"] >= 60) & (df["energy_%"] <= 80) &
    (df["speechiness_%"] < 10) &
    (df["acousticness_%"] < 30)
]

filtered_df[["track_name", "artist(s)_name", "bpm", "key", "mode", "danceability_%", "energy_%"]]
```
